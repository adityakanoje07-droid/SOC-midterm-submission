**3.**Given a lowercase character c and a non-negative integer k, output the character that is k positions after c in the alphabet. Wrap around from 'z' to 'a' if needed.

**Input Format:**

One character and one integer separated by space

**Output Format:**

One character (the shifted result)

```python
char_input, k_input = input().split()
k = int(k_input)
current_index = ord(char_input) - ord('a')
new_index = (current_index + k) % 26
result_char = chr(new_index + ord('a'))
print(result_char)
```

**4.**Keep taking integer input until you get a negative number. Then print the total number of digits (0–9) you encountered in the input before the negative number.

**Input Format:**

One integer per line until a negative number is entered.

**Output Format:**

A single integer (total count of digits seen).

```python
total_digits = 0

while True:
    user_input = input().strip()
    num = int(user_input)

    if num < 0:
        break

    total_digits += len(user_input)

print(total_digits)
```

**5.**You're given n integers. Print the sum of all numbers that are divisible by 3 or 5 but not both.

**Input Format:**

First line: n

Second line: n integers separated by space

**Output Format:**

Single integer (sum of required numbers)

```python
n = int(input())

total_sum = 0
for x in input().split():
    num = int(x)
    div_by_3 = (num % 3 == 0)
    div_by_5 = (num % 5 == 0)

    if div_by_3 != div_by_5:
        total_sum += num

print(total_sum)
```

**6.**Given a positive integer n, keep summing its digits until you're left with a single-digit number. Return that number.

**Input Format:**

A single integer

**Output Format:**

The single-digit result

```python
n_str = input().strip()

while len(n_str) > 1:
    digit_sum = 0
    for digit in n_str:
        digit_sum += int(digit)
    n_str = str(digit_sum)

print(n_str)
```
