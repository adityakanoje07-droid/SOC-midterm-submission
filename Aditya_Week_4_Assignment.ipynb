## Q1: Graph Representation and Traversal

You are given an undirected graph with `n` nodes and `m` edges. Write code to:
- Represent it using an adjacency list.
- Perform BFS and DFS traversal starting from node 0.

### Input Format:
```
n m
u1 v1
u2 v2
...
um vm
```

### Output Format:
```
BFS: node1 node2 ...
DFS: node1 node2 ...
```


````python
from collections import deque

# Input number of nodes and edges
n, m = map(int, input("Enter number of nodes and edges: ").split())

# Create adjacency list
graph = {i: [] for i in range(n)}

# Input edges
print("Enter edges:")
for _ in range(m):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)  # Undirected graph

# Sort adjacency lists for consistent traversal
for node in graph:
    graph[node].sort()

# BFS Traversal
def bfs(start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    result = []

    while queue:
        node = queue.popleft()
        result.append(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return result

# DFS Traversal
def dfs(start):
    visited = set()
    result = []

    def dfs_helper(node):
        visited.add(node)
        result.append(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_helper(neighbor)

    dfs_helper(start)
    return result

# Perform traversals
bfs_result = bfs(0)
dfs_result = dfs(0)

# Output
print("BFS:", *bfs_result)
print("DFS:", *dfs_result)
````


## Q2: Cycle Detection in Undirected Graph

Given an undirected graph, check whether it contains a cycle using DFS.

### Input:
Number of nodes and edges followed by edge list.

### Output:
`Yes` if cycle exists, otherwise `No`.


````python

n, m = map(int, input("Enter number of nodes and edges: ").split())

# Create adjacency list
graph = {i: [] for i in range(n)}

print("Enter edges:")
for _ in range(m):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

visited = set()

def dfs(node, parent):
    visited.add(node)

    for neighbor in graph[node]:
        if neighbor not in visited:
            if dfs(neighbor, node):
                return True
        elif neighbor != parent:
            return True

    return False

cycle = False

for node in range(n):
    if node not in visited:
        if dfs(node, -1):
            cycle = True
            break

print("Yes" if cycle else "No")
````


## Q3: Topological Sort (Kahn's Algorithm)

Given a **directed acyclic graph**, return any valid **topological ordering** using Kahn’s Algorithm.

### Input Format:
```
n m
u1 v1
u2 v2
...
um vm
```

### Output:
```
Topological Order: node1 node2 ...
```


````python
from collections import deque

# Input number of nodes and edges
n, m = map(int, input("Enter number of nodes and edges: ").split())

# Adjacency list and indegree array
graph = {i: [] for i in range(n)}
indegree = [0] * n

print("Enter edges (u v):")
for _ in range(m):
    u, v = map(int, input().split())
    graph[u].append(v)
    indegree[v] += 1

# Queue all nodes with indegree 0
queue = deque()

for i in range(n):
    if indegree[i] == 0:
        queue.append(i)

topo_order = []

while queue:
    node = queue.popleft()
    topo_order.append(node)

    for neighbor in graph[node]:
        indegree[neighbor] -= 1

        if indegree[neighbor] == 0:
            queue.append(neighbor)

# Check if topological sort is possible
if len(topo_order) == n:
    print("Topological Order:", *topo_order)
else:
    print("Graph contains a cycle. Topological ordering not possible.")
````


## Q4: Dijkstra’s Algorithm

Implement Dijkstra’s algorithm to find the **shortest distance from a source node (0)** to all other nodes in a graph with **non-negative edge weights**.

### Input:
```
n m
u1 v1 w1
u2 v2 w2
...
um vm wm
```

### Output:
```
0: 0
1: dist1
2: dist2
...
```


````python
import heapq

# Input number of nodes and edges
n, m = map(int, input("Enter number of nodes and edges: ").split())

# Create adjacency list
graph = {i: [] for i in range(n)}

print("Enter edges (u v w):")
for _ in range(m):
    u, v, w = map(int, input().split())

    # Undirected graph
    graph[u].append((v, w))
    graph[v].append((u, w))

# Dijkstra's Algorithm
def dijkstra(source):
    dist = [float('inf')] * n
    dist[source] = 0

    pq = [(0, source)]  # (distance, node)

    while pq:
        current_dist, node = heapq.heappop(pq)

        if current_dist > dist[node]:
            continue

        for neighbor, weight in graph[node]:
            new_dist = current_dist + weight

            if new_dist < dist[neighbor]:
                dist[neighbor] = new_dist
                heapq.heappush(pq, (new_dist, neighbor))

    return dist

# Find shortest distances from node 0
distances = dijkstra(0)

# Output
for i in range(n):
    print(f"{i}: {distances[i]}")
````


## Q5: Strongly Connected Components (Kosaraju's Algorithm)

You are given a **directed graph**. Find and print all **Strongly Connected Components** (SCCs).

### Input:
```
n m
u1 v1
u2 v2
...
um vm
```

### Output:
Print each SCC in a new line.


````python

n, m = map(int, input("Enter number of nodes and edges: ").split())

# Original graph and reversed graph
graph = {i: [] for i in range(n)}
reverse_graph = {i: [] for i in range(n)}

print("Enter directed edges (u v):")
for _ in range(m):
    u, v = map(int, input().split())
    graph[u].append(v)
    reverse_graph[v].append(u)

# Step 1: DFS to get finishing order
visited = set()
stack = []

def dfs1(node):
    visited.add(node)

    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs1(neighbor)

    stack.append(node)

for node in range(n):
    if node not in visited:
        dfs1(node)

# Step 2: DFS on reversed graph
visited.clear()

def dfs2(node, component):
    visited.add(node)
    component.append(node)

    for neighbor in reverse_graph[node]:
        if neighbor not in visited:
            dfs2(neighbor, component)

print("Strongly Connected Components:")

while stack:
    node = stack.pop()

    if node not in visited:
        component = []
        dfs2(node, component)
        print(*component)
````
