1. Create a student grade management system that stores student information and calculates statistics.

**Input Format:**

First line: number of students

Next lines: student_name subject1:grade1,subject2:grade2,...


**Output Format:**

Line 1: Dictionary of all student records

Line 2: List of students sorted by average grade (descending)

Line 3: Tuple of (highest_average, lowest_average, class_average)


**Example:**

**Input:**

3

Alice Math:85,Science:90,English:78

Bob Math:92,Science:88,English:95

Charlie Math:76,Science:82,English:80

**Output:**

{'Alice': {'Math': 85, 'Science': 90, 'English': 78}, 'Bob': {'Math': 92, 'Science': 88, 'English': 95}, 'Charlie': {'Math': 76, 'Science': 82, 'English': 80}}

['Bob', 'Alice', 'Charlie']

(91.67, 79.33, 84.44)

```python
n = int(input())

students = {}

for _ in range(n):
    data = input().split()
    name = data[0]

    grades = {}
    for item in data[1].split(","):
        subject, grade = item.split(":")
        grades[subject] = int(grade)

    students[name] = grades

print(students)

averages = {}
for name, grades in students.items():
    averages[name] = sum(grades.values()) / len(grades)

sorted_students = sorted(averages, key=averages.get, reverse=True)
print(sorted_students)

highest_average = round(max(averages.values()), 2)
lowest_average = round(min(averages.values()), 2)
class_average = round(sum(averages.values()) / len(averages), 2)

print((highest_average, lowest_average, class_average))
```

2. Analyze a text to find word frequencies, unique words, and sentence statistics.

**Input Format:**

Multiple lines of text (terminated by empty line or EOF)


**Output Format:**

Line 1: Dictionary of word frequencies (sorted by word)

Line 2: Set of unique words with length > 4

Line 3: List of sentence lengths (number of words per sentence)


**Example:**

**Input:**

The quick brown fox jumps over the lazy dog.

This is a simple sentence.

**Output:**

{'The': 1, 'This': 1, 'a': 1, 'brown': 1, 'dog': 1, 'fox': 1, 'is': 1, 'jumps': 1, 'lazy': 1, 'over': 1, 'quick': 1, 'sentence': 1, 'simple': 1, 'the': 1}

{'brown', 'jumps', 'quick', 'sentence', 'simple'}

[9, 5]

```python
text = []

while True:
    try:
        line = input()
        if line == "":
            break
        text.append(line)
    except EOFError:
        break

full_text = " ".join(text)

words = full_text.replace(".", "").split()

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

freq = dict(sorted(freq.items()))
print(freq)

unique_words = {word for word in words if len(word) > 4}
print(unique_words)

sentences = full_text.split(".")
lengths = []

for sentence in sentences:
    sentence = sentence.strip()
    if sentence:
        lengths.append(len(sentence.split()))

print(lengths)
```

3. Analyze a social network to find mutual friends, popular users, and connection statistics.

**Input Format:**

First line: number of friendships

Next lines: user1 user2 (friendship pairs)

Last line: two usernames to check mutual friends


**Output Format:**

Line 1: Dictionary of each user's friends (as sets)

Line 2: List of users sorted by number of friends (descending)

Line 3: Set of mutual friends between the two specified users

Line 4: Tuple of (total_users, total_friendships, average_friends_per_user)


**Example:**

**Input:**

5

Alice Bob

Bob Charlie

Alice Charlie

Charlie David

Bob David

Alice Charlie


**Output:**

{'Alice': {'Bob', 'Charlie'}, 'Bob': {'Alice', 'Charlie', 'David'}, 'Charlie': {'Alice', 'Bob', 'David'}, 'David': {'Bob', 'Charlie'}}

['Bob', 'Charlie', 'Alice', 'David']

{'Bob'}

(4, 5, 2.5)

```python
n = int(input())

friends_dict = {}

for _ in range(n):
    u1, u2 = input().split()

    if u1 not in friends_dict:
        friends_dict[u1] = set()
    if u2 not in friends_dict:
        friends_dict[u2] = set()

    friends_dict[u1].add(u2)
    friends_dict[u2].add(u1)

user_a, user_b = input().split()

print(friends_dict)

sorted_users = sorted(
    friends_dict.keys(),
    key=lambda x: len(friends_dict[x]),
    reverse=True
)
print(sorted_users)

mutual = friends_dict[user_a].intersection(friends_dict[user_b])
print(mutual)

total_users = len(friends_dict)
total_friendships = n
total_connections = sum(len(v) for v in friends_dict.values())
avg_friends = total_connections / total_users

print((total_users, total_friendships, avg_friends))
```

4. Manage a store inventory with operations like adding stock, processing orders, and generating reports.

**Input Format:**

First line: number of initial inventory items

Next lines: item_name:quantity:price

Following line: number of operations

Next lines: operation_type:parameters


Operations:

ADD:item_name:quantity:price

SELL:item_name:quantity

RESTOCK:item_name:quantity

CHECK:item_name


**Output Format:**

Line 1: Final inventory dictionary

Line 2: List of items that need restocking (quantity < 5)

Line 3: Set of items that were sold

Line 4: Tuple of (total_items, total_value, most_expensive_item)


**Example:**

**Input:**

3

Laptop:10:1000

Mouse:25:20

Keyboard:15:50

4

SELL:Laptop:2

ADD:Monitor:5:300

RESTOCK:Mouse:10

SELL:Keyboard:5

**Output:**

{'Laptop': (8, 1000), 'Mouse': (35, 20), 'Keyboard': (10, 50), 'Monitor': (5, 300)}

['Monitor']

{'Laptop', 'Keyboard'}

(4, 10800, 'Laptop')

```python
n = int(input())

inventory = {}

for _ in range(n):
    item, qty, price = input().split(":")
    inventory[item] = (int(qty), int(price))

m = int(input())

sold_items = set()

for _ in range(m):
    op = input().split(":")

    if op[0] == "ADD":
        item, qty, price = op[1], int(op[2]), int(op[3])
        inventory[item] = (qty, price)

    elif op[0] == "SELL":
        item, qty = op[1], int(op[2])
        if item in inventory:
            curr_qty, price = inventory[item]
            inventory[item] = (curr_qty - qty, price)
            sold_items.add(item)

    elif op[0] == "RESTOCK":
        item, qty = op[1], int(op[2])
        if item in inventory:
            curr_qty, price = inventory[item]
            inventory[item] = (curr_qty + qty, price)

    elif op[0] == "CHECK":
        item = op[1]
        if item in inventory:
            print(inventory[item])

print(inventory)

restock = sorted([item for item, (qty, price) in inventory.items() if qty <= 5])
print(restock)

print(sold_items)

total_items = len(inventory)
total_value = sum(qty * price for qty, price in inventory.values())
most_expensive_item = max(inventory, key=lambda x: inventory[x][1])

print((total_items, total_value, most_expensive_item))
```

5. Create a course scheduling system that handles prerequisites, time conflicts, and student enrollments.

**Input Format:**

First line: number of courses

Next lines: course_code:course_name:credits:time_slot:max_students:prerequisites

Following line: number of students

Next lines: student_id:preferred_courses (comma-separated)


**Output Format:**

Line 1: Dictionary of course information

Line 2: List of courses sorted by enrollment demand

Line 3: Set of courses with prerequisite conflicts

Line 4: Tuple of (total_enrollments, courses_at_capacity, average_credits_per_student)


**Example:**

**Input:**

4

CS101:Intro Programming:3:MWF9:30:

CS201:Data Structures:3:MWF10:25:CS101

MATH101:Calculus:4:TTH9:40:

PHYS101:Physics:4:TTH10:35:MATH101

3

S001:CS101,MATH101

S002:CS201,PHYS101

S003:CS101,CS201,MATH101


**Output:**

{'CS101': {'name': 'Intro Programming', 'credits': 3, 'time': 'MWF9', 'max': 30, 'prereq': set(), 'enrolled': {'S001', 'S003'}}, 'CS201': {'name':
'Data Structures', 'credits': 3, 'time': 'MWF10', 'max': 25, 'prereq': {'CS101'}, 'enrolled': {'S003'}}, 'MATH101': {'name': 'Calculus', 'credits': 4,
'time': 'TTH9', 'max': 40, 'prereq': set(), 'enrolled': {'S001', 'S003'}}, 'PHYS101': {'name': 'Physics', 'credits': 4, 'time': 'TTH10', 'max': 35,
'prereq': {'MATH101'}, 'enrolled': set()}}

['CS101', 'MATH101', 'CS201', 'PHYS101']

{'CS201', 'PHYS101'}

(5, 0, 6.67)

```python
n = int(input())

courses = {}

for _ in range(n):
    data = input().split(":")
    code, name, credits, time_slot, max_students = data[:5]

    prereq = set(data[5].split(",")) if len(data) > 5 and data[5] else set()

    courses[code] = {
        "name": name,
        "credits": int(credits),
        "time": time_slot,
        "max": int(max_students),
        "prereq": prereq,
        "enrolled": set()
    }

m = int(input())

prereq_conflicts = set()
total_credits = 0
total_enrollments = 0

for _ in range(m):
    student_id, preferred = input().split(":")
    preferred_courses = preferred.split(",")

    enrolled_courses = set()

    for course in preferred_courses:
        if course not in courses:
            continue

        prereqs = courses[course]["prereq"]

        if prereqs and not prereqs.issubset(set(preferred_courses)):
            prereq_conflicts.add(course)
            continue

        courses[course]["enrolled"].add(student_id)
        enrolled_courses.add(course)

    total_enrollments += len(enrolled_courses)

    total_credits += sum(
        courses[c]["credits"] for c in enrolled_courses
    )

print(courses)

sorted_courses = sorted(
    courses.keys(),
    key=lambda x: len(courses[x]["enrolled"]),
    reverse=True
)
print(sorted_courses)

print(prereq_conflicts)

courses_at_capacity = sum(
    1 for c in courses
    if len(courses[c]["enrolled"]) >= courses[c]["max"]
)

avg_credits = round(total_credits / m, 2)

print((total_enrollments, courses_at_capacity, avg_credits))
```

6. Build a recommendation system that analyzes purchase history and suggests products based on similarity and popularity.

**Input Format:**

First line: number of customers

Next lines: customer_id:purchased_items (comma-separated)

Following line: number of products with categories

Next lines: product_id:category:price

Last line: customer_id to generate recommendations for


**Output Format:**

Line 1: Dictionary of customer purchase profiles

Line 2: Dictionary of product categories with their average prices

Line 3: Set of customers with similar purchase patterns (Jaccard similarity > 0.3)

Line 4: List of top 5 recommended products (not already purchased)

Line 5: Tuple of (total_customers, most_popular_category, recommendation_confidence)


**Example:**

**Input:**

4

C001:P001,P002,P003

C002:P001,P004,P005

C003:P002,P003,P006

C004:P001,P002,P007

5

P001:Electronics:100

P002:Electronics:150

P003:Books:20

P004:Electronics:200

P005:Books:15

C001


**Output:**

{'C001': {'Electronics', 'Books'}, 'C002': {'Electronics', 'Books'}, 'C003': {'Electronics', 'Books'}, 'C004': {'Electronics'}}

{'Electronics': 137.5, 'Books': 17.5}

{'C002', 'C003', 'C004'}

['P004', 'P005']

(4, 'Electronics', 0.75)

```python
n = int(input())

customer_products = {}

for _ in range(n):
    cid, products = input().split(":")
    customer_products[cid] = set(products.split(","))

m = int(input())

product_info = {}
category_prices = {}

for _ in range(m):
    pid, category, price = input().split(":")
    price = float(price)

    product_info[pid] = (category, price)

    if category not in category_prices:
        category_prices[category] = []
    category_prices[category].append(price)

customer_profiles = {}

for cid, products in customer_products.items():
    categories = set()
    for p in products:
        if p in product_info:
            categories.add(product_info[p][0])
    customer_profiles[cid] = categories

print(customer_profiles)

avg_prices = {}
for category, prices in category_prices.items():
    avg_prices[category] = sum(prices) / len(prices)

print(avg_prices)

target = input()

similar_customers = set()

target_products = customer_products[target]

for cid, products in customer_products.items():
    if cid != target:
        intersection = len(target_products & products)
        union = len(target_products | products)

        if union > 0 and intersection / union > 0.3:
            similar_customers.add(cid)

print(similar_customers)

recommended = set()

for cid in similar_customers:
    recommended.update(customer_products[cid] - target_products)

recommended = sorted(recommended)[:5]
print(recommended)

category_count = {}

for products in customer_products.values():
    for p in products:
        if p in product_info:
            category = product_info[p][0]
            category_count[category] = category_count.get(category, 0) + 1

most_popular_category = max(category_count, key=category_count.get)

recommendation_confidence = round(
    len(similar_customers) / (n - 1), 2
) if n > 1 else 0

print((n, most_popular_category, recommendation_confidence))
```

7. Manage a tournament bracket system with player rankings, match results, and statistics.

**Input Format:**

First line: number of players

Next lines: player_name:initial_rating

Following line: number of matches played

Next lines: player1:player2:winner:score1:score2


**Output Format:**

Line 1: Dictionary of player statistics (wins, losses, rating)

Line 2: List of players ranked by final rating

Line 3: Set of players who beat higher-rated opponents

Line 4: Tuple of matches played by each player (as tuple pairs)

Line 5: Tournament statistics (total_matches, upsets, average_rating_change)


**Example:**

**Input:**

4

Alice:1200

Bob:1100

Charlie:1300

David:1000

3

Alice:Bob:Alice:3:1

Charlie:David:Charlie:2:0

Alice:Charlie:Charlie:1:2


**Output:**

{'Alice': {'wins': 1, 'losses': 1, 'rating': 1190}, 'Bob': {'wins': 0, 'losses': 1, 'rating': 1090}, 'Charlie': {'wins': 2, 'losses': 0, 'rating':
1320}, 'David': {'wins': 0, 'losses': 1, 'rating': 990}}

['Charlie', 'Alice', 'Bob', 'David']

set()

[('Alice', 2), ('Bob', 1), ('Charlie', 2), ('David', 1)]

(3, 0, 15.0)

```python
n = int(input())

players = {}

for _ in range(n):
    name, rating = input().split(":")
    players[name] = {
        "wins": 0,
        "losses": 0,
        "rating": int(rating)
    }

m = int(input())

upsets = set()
match_count = {player: 0 for player in players}
total_rating_change = 0

for _ in range(m):
    p1, p2, winner, s1, s2 = input().split(":")

    r1 = players[p1]["rating"]
    r2 = players[p2]["rating"]

    match_count[p1] += 1
    match_count[p2] += 1

    if winner == p1:
        loser = p2
        if r1 < r2:
            upsets.add(p1)

        players[p1]["wins"] += 1
        players[p2]["losses"] += 1

        players[p1]["rating"] += 10
        players[p2]["rating"] -= 10

    else:
        loser = p1
        if r2 < r1:
            upsets.add(p2)

        players[p2]["wins"] += 1
        players[p1]["losses"] += 1

        players[p2]["rating"] += 10
        players[p1]["rating"] -= 10

    total_rating_change += 20

print(players)

ranked_players = sorted(
    players.keys(),
    key=lambda x: players[x]["rating"],
    reverse=True
)
print(ranked_players)

print(upsets)

matches_played = [(player, match_count[player]) for player in players]
print(matches_played)

avg_rating_change = total_rating_change / (m + 1)

print((m, len(upsets), avg_rating_change))
```

8. Analyze traffic data from a smart city system to optimize routes and identify patterns.

**Input Format:**

First line: number of intersections

Next lines: intersection_id:connected_intersections:traffic_density

Following line: number of traffic events

Next lines: timestamp:intersection_id:event_type:severity

Last line: start_intersection end_intersection (for route optimization)


**Output Format:**

Line 1: Dictionary of intersection connectivity and traffic data

Line 2: List of intersections sorted by total traffic events

Line 3: Set of high-traffic intersections (density > average)

Line 4: Tuple of optimal route (intersection sequence)

Line 5: Dictionary of event statistics by type and severity


**Example:**

**Input:**

4

I001:I002,I003:45

I002:I001,I004:60

I003:I001,I004:30

I004:I002,I003:75

6

09:00:I001:accident:high

09:15:I002:congestion:medium

09:30:I001:roadwork:low

10:00:I004:accident:high

10:15:I003:congestion:low

10:30:I002:congestion:medium

I001 I004


**Output:**

{'I001': {'connections': {'I002', 'I003'}, 'density': 45, 'events': 2}, 'I002': {'connections': {'I001', 'I004'}, 'density': 60, 'events': 2}, 'I003':
{'connections': {'I001', 'I004'}, 'density': 30, 'events': 1}, 'I004': {'connections': {'I002', 'I003'}, 'density': 75, 'events': 1}}

['I001', 'I002', 'I004', 'I003']

{'I002', 'I004'}

('I001', 'I002', 'I004')

{'accident': {'high': 2}, 'congestion': {'medium': 2, 'low': 1}, 'roadwork': {'low': 1}}

```python
from collections import deque

n = int(input())

intersections = {}

for _ in range(n):
    iid, connections, density = input().split(":")
    intersections[iid] = {
        "connections": set(connections.split(",")),
        "density": int(density),
        "events": 0
    }

m = int(input())

event_stats = {}

for _ in range(m):
    timestamp, iid, event_type, severity = input().split(":")

    intersections[iid]["events"] += 1

    if event_type not in event_stats:
        event_stats[event_type] = {}

    event_stats[event_type][severity] = (
        event_stats[event_type].get(severity, 0) + 1
    )

print(intersections)

sorted_intersections = sorted(
    intersections.keys(),
    key=lambda x: intersections[x]["events"],
    reverse=True
)
print(sorted_intersections)

avg_density = sum(
    intersections[i]["density"] for i in intersections
) / len(intersections)

high_traffic = {
    i for i in intersections
    if intersections[i]["density"] > avg_density
}
print(high_traffic)

start, end = input().split()

queue = deque([[start]])
visited = {start}
route = ()

while queue:
    path = queue.popleft()
    node = path[-1]

    if node == end:
        route = tuple(path)
        break

    for neighbor in intersections[node]["connections"]:
        if neighbor not in visited:
            visited.add(neighbor)
            queue.append(path + [neighbor])

print(route)

print(event_stats)
```
